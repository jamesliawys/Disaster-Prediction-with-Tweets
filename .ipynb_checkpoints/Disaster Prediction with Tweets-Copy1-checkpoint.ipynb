```python
import gc
import pandas as pd
import string
from tensorflow.keras.callbacks import EarlyStopping

gc.collect()

class Cfg:
    # DataFrames
    train = pd.read_csv('C:/Users/User/Desktop/Jupyter/Disaster Tweets/train.csv', index_col=0)
    test = pd.read_csv('C:/Users/User/Desktop/Jupyter/Disaster Tweets/test.csv')
    sample = pd.read_csv('C:/Users/User/Desktop/Jupyter/Disaster Tweets/sample_submission.csv')
    
    # TF-IDF
    from nltk.corpus import stopwords
    feature = 'text'
    target = 'target'
    language = 'english'
    stop_words = stopwords.words(language)
    punct = string.punctuation
    
    # Test size & Seed
    seed = 42
    test_size = 0.2
    
    # DL
    model = "roberta_base_en"
    batch_size = 32
    n_epochs = 2
    lr = 1e-5
    n_classes = 2
    seq_length = 150
    metrics = ["accuracy"]
    callback = EarlyStopping(monitor='loss', patience=3)
```

```python
#EDA

df = Cfg.train

df.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = " ".join(i for i in df['text'])
wcloud = WordCloud(random_state=Cfg.seed).generate(text)
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wcloud);
```

```python
from nltk.tokenize import word_tokenize

sentence_example = df.iloc[1][Cfg.feature]
tokens = word_tokenize(sentence_example, language=Cfg.language)
tokens_no_punctuation = [i.lower() for i in tokens if i not in Cfg.punct]
tokens_no_punctuation_and_stopwords = [i for i in tokens_no_punctuation if i not in Cfg.stop_words]
gc.collect()
```

```python
def tokenize(sentence: str):
    tokens = word_tokenize(sentence, language=Cfg.language)
    tokens = [i.lower() for i in tokens if i not in Cfg.punct]
    tokens = [i for i in tokens if i not in Cfg.stop_words]
    return tokens
```

```python
%%time
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(df, test_size=Cfg.test_size, 
                                    shuffle=True, random_state=Cfg.seed)
```

```python
%%time
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

model_pipeline = Pipeline([('vectorizer', TfidfVectorizer(tokenizer=lambda x: tokenize(x))),
                           ('model', LogisticRegression(random_state=Cfg.seed))])
model_pipeline.fit(train_df[Cfg.feature], train_df[Cfg.target])
tf_idf_preds = model_pipeline.predict(val_df[Cfg.feature])
gc.collect()
```

```python
%%time
sample = Cfg.sample
sample.head()
```

```python
%%time
test = Cfg.test
```

```python
%%time
sample[Cfg.target] = model_pipeline.predict(test[Cfg.feature])
sample.to_csv('tf_idf_submission.csv', index=False)
```

```python
#Applying BERT model
!pip install -q keras-core --upgrade
!pip install -q keras-nlp --upgrade

from keras_nlp.models import RobertaClassifier, RobertaPreprocessor

preprocessor = RobertaPreprocessor.from_preset(Cfg.model, 
                                            sequence_length=Cfg.seq_length)
roberta = RobertaClassifier.from_preset(Cfg.model, 
                                  preprocessor=preprocessor, 
                                  num_classes=Cfg.n_classes)
roberta.summary()
```

```python
%%time
roberta.compile(
    loss=SparseCategoricalCrossentropy(from_logits=True),
    optimizer=Adam(Cfg.lr),
    metrics=Cfg.metrics  
)

history = roberta.fit(x=train_df[Cfg.feature],
                   y=train_df[Cfg.target],
                   batch_size=Cfg.batch_size,
                   epochs=Cfg.n_epochs, 
                   validation_data=(val_df[Cfg.feature],
                                    val_df[Cfg.target]),
                   callbacks=[Cfg.callback]
                        )
```

```python
%%time
y_pred_train = roberta.predict(train_df[Cfg.feature])
y_pred_val = roberta.predict(val_df[Cfg.feature])
```
